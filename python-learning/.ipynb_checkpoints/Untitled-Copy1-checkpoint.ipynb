```python
class Person(object):
    def __init__(self, first, last, age):
        self.first = first
        self.last = last
        self.age = age
    def full_name(self):
        return self.first + ' ' + self.last
person = Person("hu", "ziang", 20)
person.full_name()
person.newage = 21
person.newage
```
